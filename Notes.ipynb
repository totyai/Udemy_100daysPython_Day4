### Challange

Given a list of names, useing the random fucntion, pick out one from the list.

choice() is getting back a random item from the list

len() is getting back the length of list, string

```python
friends = ["Alice", "Bob", "Charlie", "David", "Emanuel"]

random_number = random.randint(0, len(friends) - 1)
print(friends[random_number])

name = random.choice(friends)
print(name)
```

## Index Error and nested lists

If you add a number that is higher then the number of items in the list, you will get the index error

```python
dirty_dozen = [
    "Strawberries",
    "Spinach",
    "Kale",
    "Nectarines",
    "Apples",
    "Grapes"]
print(dirty_dozen[10])
```

You can nest lists within list, and python will keep them separated

```python
fruits = ["Apple", "Banana", "Cherry"]
vegetables = ["Carrot", "Lettuce", "Onion"] 
dirty_dozen = [fruits, vegetables]
print(dirty_dozen)

# Since we have multiple lists inside each other. To get a hold of item from the lists, you need a matrixed indexed call back.
print(dirty_dozen[0][1]) # This will be Banana - the 0 represents the first list, fruits, and the 1 represents the second item in that list.
```
